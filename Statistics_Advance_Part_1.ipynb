```python
#Q3 What is the difference between discrete and continuous distributions?
#   Discrete Distributions-- Distributions where the random variable takes on a countable number of distinct values.
#                            Values Can be listed (like 0, 1, 2, 3...) — no in-between values.
#
#   Continuous Distributions-- Distributions where the random variable can take on any value within an interval (uncountably infinite).
#                              Values include fractions and decimals — infinitely many possibilities.
```

```python
#Q4 What are probability distribution functions (PDF)?
#   In probability theory, a Probability Distribution Function (often abbreviated as PDF) is a mathematical function that describes the likelihood of a random variable taking on a range of values.
```

```python
#Q5 How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
#   CDF (Cumulative Distribution Function)--
#     Applies to both discrete and continuous variables.
#     Describes the probability that the variable is less than or equal to a value.
#     It's a running total of probabilities up to a point.
#     Defined as: F(x)=P(X≤x)

#   PDF (Probability Density Function)--
#     Applies to both discrete and continuous variables.
#     Describes the probability that the variable is less than or equal to a value.
#     It's a running total of probabilities up to a point.
#     Defined as: F(x)=P(X≤x)
```

```python
#Q12 What is the standard normal distribution, and why is it important?
#    The standard normal distribution is a special case of the normal distribution with:
#    Mean μ=0 and Standard deviation = 1
#    Why It's Important:
#         Simplifies calculations:Probabilities and areas under the curve are tabulated in Z-tables.
#         Used in hypothesis testing, confidence intervals, and statistical inference.
```

```python
#Q13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?
#    The sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's original distribution, as long as the population has a finite mean and variance.
#    It's Critical in Statistics: Makes inference possible, Works for most real-world data, Foundation for hypothesis testing.
```

```python
#Q14 How does the Central Limit Theorem relate to the normal distribution?
#    The Central Limit Theorem (CLT) explains why the normal distribution is so widely used in statistics.
#    The CLT says: When you take large enough random samples from any population (with a finite mean and variance),
#      the distribution of the sample means will approximate a normal distribution — regardless of the population's shape.
```

```python
#Q15 What is the application of Z statistics in hypothesis testing?
#    Z-statistics are used in hypothesis testing to determine how far a sample statistic is from the population parameter, in terms of standard deviations, assuming the null hypothesis is true.
```

```python
#Q16 How do you calculate a Z-score, and what does it represent?
#    A Z-score tells you how many standard deviations a value is from the mean of a distribution.
#     Z = X-μ/=σ

#     Z = 0 → the value is exactly at the mean
#     Z > 0 → the value is above the mean
#     Z < 0 → the value is below the mean
```

```python
#Q17 What are point estimates and interval estimates in statistics?
#    Point Estimate-- A point estimate is a single value used to estimate an unknown population parameter.
#    Interval Estimate-- An interval estimate gives a range of values, along with a level of confidence, within which the population parameter is likely to lie.
```

```python
#Q18 What is the significance of confidence intervals in statistical analysis?
#    A confidence interval (CI) provides a range of values that is likely to contain the true population parameter, along with a level of confidence (e.g., 95%).
```

```python
#Q19 What is the relationship between a Z-score and a confidence interval?
#    A Z-score is used to calculate confidence intervals when the population standard deviation is known and the sampling distribution is normal (or sample size is large).
#    A confidence interval is calculated using a Z-score that corresponds to the desired confidence level.
```

```python
#Q20 How are Z-scores used to compare different distributions?
#    Z-scores standardize values from different distributions, allowing you to compare them on the same scale, even if the original units or scales are different.
```

```python
#Q21  What are the assumptions for applying the Central Limit Theorem?
#     To correctly apply the Central Limit Theorem, the following assumptions must generally be met:
#      Independent Observations, Identically Distributed, Finite Mean and Variance,  Sample Size is Sufficiently Large
```

```python
#Q22 What is the concept of expected value in a probability distribution?
#    The expected value (also called mean or expectation) is the average outcome you’d expect if an experiment or process were repeated many times.
#    It’s the long-run average value of a random variable.
```

```python
#Q2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution (e.g., 1 to 6 for a fair die)
low, high = 1, 7  # upper bound is exclusive in scipy's randint

# Create a discrete uniform distribution
dist = randint(low, high)

# Define the values (support) of the distribution
x = np.arange(low, high)

# Get the PMF values
pmf_values = dist.pmf(x)

# Plot the PMF
# Remove 'use_line_collection=True' since it's not supported in older versions.
# You might see a performance difference, especially for large datasets.
plt.stem(x, pmf_values, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python
#Q9 Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    # Plot the Z-scores
    plt.hist(z_scores, bins=30, density=True, color='orange', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

    return z_scores

data = np.random.normal(50, 10, 1000)  # Simulated dataset
z_scores = calculate_and_plot_zscores(data)
```

```python
#Q10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000
sample_size = 30
num_samples = 1000

# Generate non-normal population data (exponential distribution)
population = np.random.exponential(scale=2.0, size=population_size)

# Collect sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Central Limit Theorem Demo\nDistribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#Q Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_standard_normal_with_z(z_score):
    # Create x values and corresponding PDF values
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    # Plot the standard normal curve
    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, y, where=(x <= z_score), color='lightblue', alpha=0.5)
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example
data_point = 75
population_mean = 70
population_std = 5

z = calculate_z_score(data_point, population_mean, population_std)
print(f"Z-score: {z:.2f}")

# Plot the Z-score on standard normal distribution
plot_standard_normal_with_z(z)
```

```python
#Q Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

# Sample data (e.g., test scores from a sample of students)
sample = np.array([72, 75, 78, 71, 69, 74, 73, 70, 76, 77])

# Population parameters (assumed known)
population_mean = 70
population_std = 5

# Step 1: Calculate sample mean and size
sample_mean = np.mean(sample)
n = len(sample)

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))  # Step 2: Calculate Z-statistic

p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # Step 3: Calculate p-value for two-tailed test

# Step 4: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.2f}")
print(f"P-value = {p_value:.4f}")

# Step 5: Conclusion at α = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#Q Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

# Sample dataset (e.g., heights, scores, etc.)
data = np.array([72, 75, 78, 71, 69, 74, 73, 70, 76, 77])

# Confidence level
confidence = 0.95

# Calculate statistics
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# Margin of error
margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * std_err

# Confidence interval
ci_lower = mean - margin
ci_upper = mean + margin

print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate normal data
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Calculate statistics
mean = np.mean(data)
sem = stats.sem(data)  # Standard error
confidence = 0.95
df = len(data) - 1

# Step 3: Calculate margin of error
t_critical = stats.t.ppf((1 + confidence) / 2, df)
margin_of_error = t_critical * sem

# Step 4: Confidence Interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Step 5: Print and interpret
print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 6: Optional: Visualize
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label='Sample Mean')
plt.axvspan(ci_lower, ci_upper, color='yellow', alpha=0.3, label='95% Confidence Interval')
plt.title('Histogram with 95% Confidence Interval for Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    # Generate data from normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # By default, this is excess kurtosis

    print(f"Skewness: {data_skewness:.4f} (≈ 0 for symmetric distribution)")
    print(f"Kurtosis: {data_kurtosis:.4f} (≈ 0 for normal distribution)")

    return data_skewness, data_kurtosis

# Example usage
analyze_normal_distribution_skew_kurtosis()
```
